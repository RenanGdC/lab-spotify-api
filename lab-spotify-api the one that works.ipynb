```python
import spotipy
import pandas as pd
import json
from spotipy.oauth2 import SpotifyClientCredentials
from config import *
```

```python
#Initialize SpotiPy with user credentias # 
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= client_id,
                                                           client_secret= client_secret))
```

```python
not_100 = pd.read_csv("not_hot_songs.csv")
```

```python
not_100.head()
```

```python
hot_100= pd.read_csv("hot_songs.csv")
hot_100.head()
```

```python
not_100['title']=not_100[['title']].apply(lambda x: x.str.lower())
not_100['artist']=not_100[['artist']].apply(lambda x: x.str.lower())
not_100.head()
```

```python
hot_100['title']=hot_100[['title']].apply(lambda x: x.str.lower())
hot_100['artist']=hot_100[['artist']].apply(lambda x: x.str.lower())
hot_100.head()
```

```python
def search_song(title, artist, limit=1):
    try:
        result = sp.search(q=title + artist, limit=limit)
        return result["tracks"]["items"][0]["uri"]
    except:
        print('none')
```

```python
search_song('mask off', 'future', limit=1)
```

```python
title_top_100 = hot_100['title'].to_list()
artist_top_100 = hot_100['artist'].to_list()
```

```python
uris = []
for i in range(len(title_top_100)):
    uris.append(search_song(title_top_100[i], artist_top_100[i], limit=1))
```

```python
hot_100['uri'] = uris
hot_100
```

```python
def get_audio_featurs(df):
    output = pd.DataFrame()
    for i in range(len(df['uri'])):
            my_dict = sp.audio_features(df.loc[i,['uri']])[0]
            #print(my_dict)
            my_dict_i = {key : my_dict[key] for key in list(my_dict.keys())}
            my_dict_i['title'] = df['title'][i]
            my_dict_i['artist'] = df['artist'][i]
            output = output.append(my_dict_i, ignore_index=True)
    return output
```

```python
get_audio_featurs(hot_100)
```
